```python
import os
import numpy as np

import tensorflow as tf
from matplotlib import pyplot as plt
DATA_PATH = '/content/drive/My Drive/Machine Learning/GanProjcet'
generator_path = os.path.join(DATA_PATH,"face_generator2.h5")
generator= tf.keras.models.load_model(generator_path)
SEED_SIZE = 50
```

```python
generator.summary()
```

```python
num_images = 10
noise = np.random.normal(0, 1, size=[num_images, 50])
generated_images = generator.predict(noise)

# Display the generated images
fig, axs = plt.subplots(1, num_images, figsize=(10, 10))
for i in range(num_images):
    axs[i].imshow(generated_images[i, :, :, :], cmap='gray')
    axs[i].axis('off')
plt.show()
```
